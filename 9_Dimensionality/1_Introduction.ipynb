# Curse of dimensionality 
Many machine learning problems have dataset with millions of features which can make training too slow and harder to find a good solution.  
This is called _curse of dimensionality_


Though it is often possible to reduce the number of features in any real world dataset without losign important information  
For ex: MNIST dataset, we can remove border pixels as they are always white and highly unimportant


Dimensionality reduction mostly reduces training time, only sometimes may remove noise and make model perfom better.

# Dimensionality Reduction

There are two main approaches for dimensionality reduction are: ___Projection___ and ___Manifold learning___

## Projection
Many real world datasets do not have data points spread out uniformly, as a result all the training instances can be projected in a much lower dimensional _subspace_ of a high dimensional space.

For example, below is a 3D dataset, there exists a 2D subspace around all the training instances. if we take projection of all those instances perpendicularly on the subpaces (as shown with small arrows).

After projection, the 3D dataset can be converted into 2D as shown below

This is not always the case with complex datasets, such as _swiss roll toy dataset_, if we take simple projection in a 2D subspace, it will not correctly mimic the 3D dataset.  
In this case we have to unroll the dataset to reduce its dimension

On the left, we took direct projection which does not mimic original dataset.  
The correct method is to unroll the dataset to get 2D _manifold_

## Manifold learning
- For 3D, a 2D manifold is a shape that can be bent or twisted to a 3D dataset.
- A d-dimensional manifold is a part of n-dimensional (d<n) hyperplane (the n-dimensional hyperplane resembles that d-dimensional hypeplane)   
ex:  the Swiss roll, d = 2 and n = 3: it
locally resembles a 2D plane, but it is rolled in the third dimension.

__what is a manifold?__  
A manifold is a smooth continuous surface that exists within a higher dimensional space    

__What is manifold learning?__   
A method of dimensionality reduction where we find a lower dimensional manifold where all the training instances lie, of a higher dimensional hyperplane.


__What is manifold hypothesis?__  
It states that most real world datasets of high dimensions exists close to a lower dimensional manifold.


Note- Reducing dimensionality does not always mean a simple ML model, it only makes the algorithm faster (only sometimes may reduce noise)
