___Pipeline___: Sequence of steps to build a ML model completely from raw data to deployment,
i.e Complete workflow to build a ML model  
In a pipeline, each component is _asynchronous_ i.e each component works independently, it takes data, process it, spit out output and next component takes it to further process it and spit output and so on...  
This makes the architechture quite robust because even if any component doesnt work, the model can still work for some time.  
Pipelines also help teams to divide work

___Performance measure___ :  
Evaluating performance measure is a way of measuring accuracy,errors and how well a model is working

For a regression problem following performance measures are used:
RMSE (Root mean square error), MSE (Mean squared error), MAE (Mean absolute error), R squared error

### RSME (root mean squared error)
$$ 
RSME(X,h) = \sqrt{\frac{1}{m}\sum_{i=1}^{m}(\hat{y^{(i)}}-y^{i})^2}
$$

### MAE(mean absolute error)

$$ 
MAE(X,h) = \frac{1}{m}\sum_{i=1}^{m}|\hat{y^{(i)}}-y^{i}|
$$

### MSE (mean squared error)
$$
MSE(X,h) = \frac{1}{m}\sum_{i=1}^{m}(\hat{y^{(i)}}-y^{i})^2
$$

### _Notations_
- $m$: number of instance in dataset
- $x^{(i)}$: vector containing all feature values of ith instance
- $y^{(i)}$: label of ith instance
- $X$: Matrix containing all feature values of all instances (except labels)
- $\hat{y^{(i)}}$: Predicted label of ith instance
- $h$: Hypothesis function 
$$
h(x^{(i)}) = \hat{y^{(i)}}
$$
